# 3. Create a 'loyalty flag'

### 3.1 Create a new column that calculates the max order number for each customer


```python
# Create a new column that calculates the max order number for each customer

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check output

ords_prods_merge.head(15)
```

### 3.2 Create a loyalty flag based on the max_order column, using the loc() function

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'Loyalty_flag'] = 'New Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] > 10) & (ords_prods_merge['max_order'] <= 40), 'Loyalty_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'Loyalty_flag'] = 'Loyal Customer'
```

```python
# Checking the values count for the Loyalty flag

ords_prods_merge['Loyalty_flag'].value_counts(dropna = False)
```

The results match the ones from the exercise practice.

```python
# Check output with new column

ords_prods_merge.head()
```

```python
# Drop 'Loyalty flag' to keep output clean

ords_prods_merge = ords_prods_merge.drop(columns = ['Loyalty flag'])
```

# 4. Explore the difference between the spending habits of the three types of customers

```python
# Group data based on the new loyalty flag and calculate mean, mix, and max for price

ords_prods_merge.groupby('Loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

We are trying to determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers. The above output suggests that loyal customers purchase items that on average are cheaper than the items purchased by new customers and regular customers.  

# 5. Create a 'spending flag'


 - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
 - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Create new column for average spend per user

ords_prods_merge['average_spend_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check new column output

ords_prods_merge.head(20)
```

```python
# Derive new spending flag column using the loc() function

ords_prods_merge.loc[ords_prods_merge['average_spend_user'] < 10, 'Spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend_user'] >= 10, 'Spending_flag'] = 'High Spender'
```

```python
# Check value count for spending flag

ords_prods_merge['Spending_flag'].value_counts(dropna = False)
```

```python
# Check dataframe output with the new spending flag column

ords_prods_merge.head(20)
```

# 6. Create 'order frequency flag'

- If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
- If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Create a new column showing the median of 'days_since_prior_order' for each customer (i.e. grouped by 'user_id')

ords_prods_merge['median_order_freq_user'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output

ords_prods_merge.head(20)
```

```python
# Derive new order frequency flag column using the loc() function

ords_prods_merge.loc[ords_prods_merge['median_order_freq_user'] > 20, 'Order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_freq_user'] <= 20) & (ords_prods_merge['median_order_freq_user'] > 10), 'Order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_freq_user'] <= 10, 'Order_frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts for new column

ords_prods_merge['Order_frequency_flag'].value_counts(dropna = False)
```

```python
# Checking dataframe output with the new column

ords_prods_merge.head(20)
```
