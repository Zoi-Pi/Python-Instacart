## 02.1 User-defined functions

```python
# Define the price_label function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the price_label function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# 03. Task

```python
# New client request for the two busiest and least busy days of the week - flag

result2 = []

for value in ords_prods_merge['orders_day_of_the_week']:
    if value == 0:
        result2.append("Busiest days")
    elif value == 1:
        result2.append("Busiest days")
    elif value == 4:
        result2.append("Least busy")
    elif value == 3:
        result2.append("Least busy")
    else:
        result2.append("Regularly busy")
```

```python
# Check result2 list
result2
```

```python
# Add result2 list as a column to ords_prods_merge

ords_prods_merge['busiest_days'] = result2
```

```python
# Check Frequency for 'busiest_days'
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Frequency of 'order_hour_of_day'

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

As the day has 24 hours, and there are three periods needs, I will split this up in periods of 8.

```python
# Using for-loop to create three periods of the 24-hour day

result3 = []

for value in ords_prods_merge['order_hour_of_day']:
    if value in range(9, 16):
        result3.append("Most orders")
    elif value in range(17, 22):
        result3.append("Average orders")
    elif value in range(7,8):
        result3.append("Average orders")
    else:
        result3.append("Fewest orders")
```

```python
result3
```

```python
# Add result3 list as a column to ords_prods_merge

ords_prods_merge['busiest_period_of_day'] = result3
```

```python
# Checking frequency for 'busiest_period_of_day'

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
